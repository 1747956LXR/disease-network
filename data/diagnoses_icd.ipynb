```python
import pandas as pd 
data_path = './mimic-iii/'
diagnoses_icd = pd.DataFrame(pd.read_csv(data_path + 'diagnoses_icd.csv'))
```

```python
diagnoses_icd
```

```python
diagnoses_icd = diagnoses_icd.loc[:, ["hadm_id", "icd9_code"]]
```

```python
diagnoses_icd = diagnoses_icd.groupby('hadm_id')['icd9_code'].apply(' '.join).reset_index()
diagnoses_icd
```

```python
def primary(codes: str) -> str:
    lst = codes.split()
    lst = [code[:-2].lstrip('0') for code in lst]
    return max(lst, key=lst.count)

codes = "570 07030 07054 30401 2875 2760"
primary(codes)
```

```python
diagnoses_icd["primary"] = diagnoses_icd["icd9_code"].map(primary)
```

```python
diagnoses_icd
```

```python
diagnoses_icd[["hadm_id", "primary"]].to_csv("primary.csv", index=None)
```
